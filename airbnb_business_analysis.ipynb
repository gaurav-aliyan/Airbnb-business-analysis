```python
import pandas as pd 
listings = pd.read_csv("listings.csv", engine = "python", on_bad_lines = "skip")
rates = pd.read_csv("past_rates.csv", engine = "python", on_bad_lines = "skip")
```

# IMPORTANT COLUMNS FROM LISTINGS

```python
important_cols = [
    'listing_id',
    'city',
    'country',
    'room_type',
    'listing_type',
    'guests',
    'bedrooms',
    'baths',
    'superhost',
    'rating_overall',
    'rating_cleanliness',
    'ttm_revenue_native',
    'ttm_occupancy'
]

listings = listings[important_cols]
```

```python
print(listings.head())
```

```python
print(listings.columns.tolist())
```

```python
print(listings.isnull().sum())
```

# Filling missing columns

```python
listings.loc[:, 'room_type'] = listings['room_type'].fillna(listings['room_type'].mode()[0])
listings.loc[:, 'listing_type'] = listings['listing_type'].fillna(listings['listing_type'].mode()[0])
listings.loc[:, 'superhost'] = listings['superhost'].fillna(listings['superhost'].mode()[0])

listings.loc[:, 'guests'] = listings['guests'].fillna(listings['guests'].median())
listings.loc[:, 'bedrooms'] = listings['bedrooms'].fillna(listings['bedrooms'].median())
listings.loc[:, 'baths'] = listings['baths'].fillna(listings['baths'].median())

listings.loc[:, 'rating_overall'] = listings['rating_overall'].fillna(listings['rating_overall'].median())
listings.loc[:, 'rating_cleanliness'] = listings['rating_cleanliness'].fillna(listings['rating_cleanliness'].median())
```

```python
print(listings.isnull().sum())
```

# dropping the column (KPI)

```python
listings = listings.dropna(subset=['ttm_revenue_native','ttm_occupancy'])
```

# checking for duplicates

```python
print(listings.duplicated().sum())
```

```python
print(listings.isnull().sum())
```

# grouping according to number of bedrooms

```python
def bedroom_group(x):
    if x<=2:
        return 'small'
    elif x<=4:
        return 'medium'
    else:
        return 'large'

listings['bedroom_groups'] = listings['bedrooms'].apply(bedroom_group)
```

```python
listings.groupby('bedroom_groups')['ttm_revenue_native'].mean()
```

```python
print(rates.columns)
```

# merging the two csv files

```python
listings['listing_id'] = listings['listing_id'].astype(str)
rates['listing_id'] = rates['listing_id'].astype(str)
```

```python
merged = pd.merge(listings, rates, on = 'listing_id')
```

```python
print(merged.columns.tolist())
```

```python
merged = merged.rename(columns={
    'city_x': 'city',
    'country_x': 'country'
})

merged = merged.drop(columns=['city_y', 'country_y'])
```

```python
print(merged.isnull().sum())
```

# handling missing values

```python
numeric_cols = [
    'booked_rate_avg',
    'booking_lead_time_avg',
    'length_of_stay_avg',
    'min_nights_avg'
]

for col in numeric_cols:
    merged[col] = merged[col].fillna(merged[col].median())
```

```python
print(merged.isnull().sum())
```

```python
print(merged.duplicated().sum())
```

# Date Conversion

```python
merged['date'] = pd.to_datetime(merged['date'])
```

```python
print(merged.head())
```

# occupancy according to Months

```python
merged.groupby('date')['occupancy'].mean()
```

# Monthly Revenue Trend

```python
monthly_revenue = merged.groupby('date')['revenue'].mean().sort_values(ascending=False)
print(monthly_revenue.head(12))
```

# Correlation between Price and Occupancy

```python
print(merged[['rate_avg','occupancy']].corr())
```

# Vacancy By City

```python
vacancy = merged.groupby('city')['vacant_days'].mean().sort_values(ascending=False)
print(vacancy.head(10))
```
